# 燃弧特征提取
计算单次操作中的燃弧相角、时长、能量、电量、功率（由于后期样本发生劣化，燃弧时间偏长，因此只取前期样本）
- 燃弧相角：
- 燃弧时长(arcing_duration)：$t$ 
- 燃弧能量(arcing_energy)：$\int i(t)u(t)dt$
- 燃弧电量(arcing_electricity)：$\int i(t)^2dt$
- 燃弧功率(arcing_power)：$\frac{\int i(t)u(t)dt}{t}$

```python
# 滑动平均
def movingAverage(sequence, window_size):
    window = np.ones(int(window_size)) / float(window_size)
    return np.convolve(sequence, window, 'same')
```

```python
def dataSmooth(data, smooth_win_size):
    data_smooth = pd.DataFrame()
    for col in data.columns:
        data_smooth[col] = movingAverage(data[col], smooth_win_size)
    return data_smooth
```

#### 拟合正弦函数

```python
from scipy.optimize import leastsq
import warnings
# 根据x和参数输出sin函数值
def calSinByParams(x, params):
    A, f, theta_x, theta_y = params 
    return A * np.sin(2 * np.pi * f * x + theta_x) + theta_y

def calFittingResidual(params, x, y_real):
    return y_real - calSinByParams(x, params)

def getOptFittingParams(data):
    '''
    params_init:初始化拟合参数 
        params_init[0]: A 振幅
        params_init[1]: f 频率
        params_init[2]: theta_x 偏移
        params_init[3]: theta_y 
    '''

    # 待拟合数据
    X = np.arange(len(data))
    Y = np.array(data)
    
    # 初始化参数
    Y_sorted = sorted(Y)
    n = min(int(len(Y)/100), 10)
    A = (np.mean(Y_sorted[-n:]) - np.mean(Y_sorted[:n])) / 2
    
    params_init = [A, CONST_PARAMS_DICT['frequency'], 0, 0]
    
    # 最小二乘得到拟合最优解
    #params = [params_init['amplitude'], params_init['frequency'], params_init['theta_x'], params_init['theta_y']]
    
    with warnings.catch_warnings():
        try:
            params_opt, _ = leastsq(calFittingResidual, params_init, args = (X, Y), maxfev=5000)
        except Warning as e:
            print(e)
            params_opt = params_init
    
    return params_opt
```

#### 计算开关开闸范围、燃弧范围

```python
# 计算燃弧区间
def findArcingRange(U_data, I_data):
    arcing_range = [-1, -1]
    
    n_points = 10
    data_len = len(U_data)
    # 确定燃弧结束位置：电流为0时表示燃弧结束（结束点后n_points对应的电流均小于电流阈值，结束点前n_points对应的电流均大于电流阈值）
    for i in range(data_len - n_points - 1, 0, -1):
        #                        #时刻在[i - n_points,i]范围内的电流都大于0                                                         # 时刻在[i + 1, i + n_points]范围内的电流都接近0
        if i - n_points >= 0 and all([1 if np.abs(Ii) > CONST_PARAMS_DICT['I_eps'] else 0 for Ii in I_data[i - n_points : i]]) and all([1 if np.abs(Ii) < CONST_PARAMS_DICT['I_eps'] else 0 for Ii in I_data[i + 1 : i + n_points]]):
            arcing_range[1] = i
            break
        
     # 确定燃弧开始位置
    arcing_range[0] = arcing_range[1]
    for i in range(arcing_range[1], 0, -1):
        if np.abs(U_data[i]) < CONST_PARAMS_DICT['U_eps']:
            if i - n_points > 0 and all([1 if np.abs(Ui) < CONST_PARAMS_DICT['U_eps'] else 0 for Ui in U_data[i - n_points : i]]):#all(U_data[i - n_points : i]):
                arcing_range[0] = i
                break
        
    return arcing_range

# 计算燃弧相角
def calArcingPhaseAngle(U_data, I_data, arcing_range):
    if arcing_range[1]  == -1:
        return 0
    phase_angle_range = arcing_range
    n_points = 10
    data_len = len(U_data)
    #确定合闸相角计算的结束点
    for i in range(arcing_range[1], data_len):
        if U_data[i] == 0 and i + 5 + n_points <= data_len and all([1 if Ui > 0 else 0 for Ui in U_data[i + 5 : i + 5 + n_points]]):
            phase_angle_range[1] = i
            break
    phase_angle = max(min(2 * np.pi, 2 * np.pi / 200 * (phase_angle_range[1] - phase_angle_range[0])), 0)
    return phase_angle

    
'''
# 计算开闸区间
def calCloseVoltageFittingParams(U_part):
    params_init_close = [0, CONST_PARAMS_DICT['frequency'], 0, 0]
    #{'amplitude': 0, 'frequency': CONST_PARAMS_DICT['frequency'], 'theta_x': 0, 'theta_y': 0}
    params_opt_close = getOptFittingParams(U_part, params_init_close)
    return params_opt_close

def findAbnormalPoints(U_data, I_data, arcing_start_loc):
    end_loc = arcing_start_loc - int(1 / CONST_PARAMS_DICT['frequency'] * 2)
    start_loc = end_loc - int(1 / CONST_PARAMS_DICT['frequency'] * 2)
    U_part = np.array(U_data[start_loc : end_loc])
    params_opt_close = calCloseVoltageFittingParams(U_part)
    #print(params_opt_close)
    
    U_part = U_data[start_loc : arcing_start_loc]
    U_real = np.array(U_part)
    U_pred = calSinByParams(np.arange(len(U_real)), params_opt_close)
    #print('U_real:',U_real)
    #print('U_pred:', U_pred)
    
    U_residual = np.abs(U_real - U_pred)
    #U_residual = pd.DataFrame(np.abs(U_real - U_pred), index = U_part.index)
    residual_threshold = 15 #2 * np.abs(params_opt_close[0])
    #print(U_residual)
    
    abnormal_points_loc = []
    for i in range(len(U_residual) - 1, -1, -1):
        if U_residual[i] > residual_threshold:
            abnormal_points_loc += [i + start_loc]
            #print(U_residual.loc[idx], idx)
    #print('abnormal_points_loc',abnormal_points_loc)        
    return abnormal_points_loc
'''
print()
```

### 燃弧特征提取

```python
# 获得燃弧特征
def extractOnePhaseArcingFeatures(U_data, I_data):
    # 计算燃弧范围
    arcing_range = findArcingRange(U_data, I_data)
    # 燃弧时长
    arcing_duration = max(0, arcing_range[1] - arcing_range[0])
    if arcing_duration > 2000:
        arcing_duration = 0
    
    if arcing_range[1] == -1:
        print(arcing_range)
    
    # 计算燃弧相角
    arcing_phase_angle = calArcingPhaseAngle(U_data, I_data, arcing_range)
    
    # 燃弧能量
    U_arcing = np.array(U_data[arcing_range[0] : arcing_range[1]])
    I_arcing = np.array(I_data[arcing_range[0] : arcing_range[1]])
    arcing_energy = 0 if arcing_duration <= 0 else np.sum([np.abs(U_arcing[i] * I_arcing[i]) for i in range(arcing_duration)])
    
    # 燃弧电量
    arcing_electricity = 0 if arcing_duration <= 0 else np.sum([I_arcing[i] ** 2 for i in range(arcing_duration)])
    
    # 燃弧功率
    arcing_power = arcing_energy / (arcing_duration + 1)
    
    
    # 计算开闸前的异常点位置
    #abnormal_points_loc = findAbnormalPoints(U_data, I_data, arcing_range[0])
    # 异常点最早出现的位置
    #abnormal_start = abnormal_points_loc[-1] if abnormal_points_loc else arcing_range[0]
    
    #open_duration = arcing_range[1] - abnormal_start
    
    #tot_ftrs = [abnormal_start, open_duration] + arcing_range + [arcing_duration, arcing_energy, arcing_electricity, arcing_power] 
    #tot_arcing_ftrs = arcing_range + [arcing_phase_angle, arcing_duration, arcing_energy, arcing_electricity, arcing_power] 
    #return tot_arcing_ftrs
    arcing_ftrs_dict = {'arcing_start': arcing_range[0], 'arcing_end': arcing_range[1], 
                        'arcing_phase_angle': arcing_phase_angle, 'arcing_duration': arcing_duration, 
                        'arcing_energy': arcing_energy, 'arcing_electricity': arcing_electricity, 'arcing_power': arcing_power}
    return arcing_ftrs_dict

# 提取合闸后稳态特征
def extractOnePhaseTouchFeatures(U_data, I_data, arcing_start_loc):
    # 设置波形截取的起止位置
    end_loc = arcing_start_loc - int(1 / CONST_PARAMS_DICT['frequency'] * 2)
    start_loc = end_loc - int(1 / CONST_PARAMS_DICT['frequency'] * 5)
    
    # 获取接触时的电流、电压波形
    U_part = np.array(U_data[start_loc : end_loc])
    I_part = np.array(I_data[start_loc : end_loc])
    
    #拟合波形，得到接触电流、接触电压、接触电阻
    #params_init = [0, CONST_PARAMS_DICT['frequency'], 0, 0]
    U_params = getOptFittingParams(U_part)#, params_init)
    I_params = getOptFittingParams(I_part)#, params_init)
    
    touch_ftrs_dict = {'touch_I': np.abs(I_params[0]), 'touch_U': np.abs(U_params[0]), 'touch_R': np.abs(U_params[0] / (I_params[0] + 0.01)) }
    return touch_ftrs_dict
    
# 提取单相特征
def extractOnePhaseFeatures(U_data, I_data):
    arcing_ftrs_dict = extractOnePhaseArcingFeatures(U_data, I_data)
    touch_ftrs_dict = extractOnePhaseTouchFeatures(U_data, I_data, arcing_ftrs_dict['arcing_start'])
    
    arcing_ftrs_dict.update(touch_ftrs_dict) #合并字典
    return arcing_ftrs_dict
```

```python
# 提取三相特征
three_phase = ['A', 'B', 'C']
arcing_ftr_columns = ['arcing_' + ftr for ftr in ['start', 'end', 'phase_angle','duration', 'energy', 'electricity', 'power']]
touch_ftr_columns = ['touch_' + ftr for ftr in ['I', 'U', 'R']]
one_phase_ftr_columns = arcing_ftr_columns + touch_ftr_columns
three_phase_ftr_columns = [ phase +'_' + ftr for phase in three_phase for ftr in one_phase_ftr_columns]

def extract3phaseFeatures(operation_data):
    machine_info_df = pd.DataFrame(columns = three_phase_ftr_columns) 
    #print(three_phase_ftr_columns)
    t_start = time.time()
    for i, tmp_operation_data in enumerate(operation_data):
        if i % 100 == 0:
            print(i)
            #print(machine_info_df)
        tmp_ftrs = []
        for tmp_phase in three_phase:
            U_data = tmp_operation_data['U' + tmp_phase]
            I_data = tmp_operation_data['I' + tmp_phase]
            tmp_ftrs_dict = extractOnePhaseFeatures(U_data, I_data)
            tmp_ftrs += [tmp_ftrs_dict[col] for col in one_phase_ftr_columns]
            
        #if tmp_ftrs[3] < 0 or tmp_ftrs[9] < 0 or tmp_ftrs[15] < 0:
        #    print(i, tmp_ftrs)
        machine_info_df = machine_info_df.append(pd.DataFrame([tmp_ftrs], columns = three_phase_ftr_columns), ignore_index = True)
        
    t_end = time.time()
    return machine_info_df
```
