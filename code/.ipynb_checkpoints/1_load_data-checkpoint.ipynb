```python
# 获得有效的生命周期
def getValidOperationRange(machine_files):
    operation_times = [ int(re.match( '(.*)年(.*)月(.*)日(.*)时(.*)分(.*)秒 第(.*)次.xlsx', machine_file).group(7)) for machine_file in machine_files]
    max_loc = (0, 0)
    cur_loc = max_loc
    for i in range(len(operation_times)):
        if i > 0:
            if operation_times[i] - operation_times[i - 1] != 1:
                cur_loc = (i, i)
            else:
                cur_loc = (cur_loc[0], i)
            if max_loc[1] - max_loc[0] < cur_loc[1] - cur_loc[0]:
                max_loc = cur_loc
    return max_loc
```

```python
# 滑动平均
def movingAverage(sequence, window_size):
    window = np.ones(int(window_size)) / float(window_size)
    return np.convolve(sequence, window, 'same')
```

#### 拟合正弦函数

```python
from scipy.optimize import leastsq

def calSinByParams(x, params):
    A, f, theta = params 
    return A * np.sin(2 * np.pi * f * x + theta)

def calFittingResidual(params, x, y_real):
    return y_real - calSinByParams(x, params)

def getOptFittingParams(U_data):
    '''
    初始化拟合参数params: 
        params[0]: A 振幅
        params[1]: f 频率
        params[2]: theta 偏移
    '''
    # 初始化拟合参数
    params_init = [421 / np.sqrt(3) *np.sqrt(2), 1/200, 0]

    # 待拟合数据
    X = np.arange(len(U_data))
    Y = np.array(U_data)
    
    # 最小二乘得到拟合最优解
    params_opt, _ = leastsq(calFittingResidual, params_init, args = (X, Y))
    
    return params_opt
```

```python
# 确定合闸位置（电压开始为0的时间）
def calCloseTime(U_real):
    
    # 设定残差阈值
    residual_threshold = 100
    n_points = 10
    
    # 拟合曲线
    U_part = U_real[:const_params_dict['period'] * 3]
    params_opt = getOptFittingParams(U_part)
    U_pred = calSinByParams(np.arange(len(U_real)), params_opt)
    U_residual = U_real - U_pred
    
    # 寻找合闸位置
    for i in range(len(U_real)):
        if U_residual[i] > residual_threshold and i + n_points < len(U_real) and all([1 if np.abs(Ui) < const_params_dict['U_eps'] else 0 for Ui in U_real[i : i+ n_points]]):
            return i
    return -1
```

```python
# 确定燃弧范围（开始：电压开始上升；结束：电流下降为0）
def findArcingRange(U_data, I_data, idx_close):
    arcing_range = [-1, -1]
    n_points = 20
    
    # 确定燃弧结束位置
    data_len = len(U_data)
    for i in range(data_len - 1, idx_close, -1):
        if np.abs(I_data[i]) > const_params_dict['I_eps']:
            if i - n_points > 0 and all([1 if np.abs(Ii) > const_params_dict['I_eps'] else 0 for Ii in I_data[i - n_points : i]]): 
                #print(i, const_params_dict['I_eps'])
                arcing_range[1] = i
                break 
                
    # 确定燃弧开始位置
    for i in range(arcing_range[1], idx_close, -1):
        if np.abs(U_data[i]) < const_params_dict['U_eps']:
            if i - n_points > 0 and all([1 if np.abs(Ui) < const_params_dict['U_eps'] else 0 for Ui in U_data[i - n_points : i]]):#all(U_data[i - n_points : i]):
                arcing_range[0] = i
                break
    return arcing_range
```

```python
const_params_dict = {
    'period': 200, 
    'U_test': 424,
    'I_test': 228,
    'U_eps': 20,
    'I_eps': 5,
}
def extractSinglePhaseFeatures(U_data, I_data):
    '''
    获得燃弧特征
    '''
    # 确定合闸时间
    idx_close = calCloseTime(U_data)
    
    # 计算燃弧区间
    arcing_range = findArcingRange(U_data, I_data, idx_close)
    #print(idx_close, arcing_range)
    
    # 燃弧时长
    arcing_duration = arcing_range[1] - arcing_range[0]
    
    # 燃弧能量
    U_arcing = np.array(U_data[arcing_range[0] : arcing_range[1]])
    I_arcing = np.array(I_data[arcing_range[0] : arcing_range[1]])
    arcing_energy = 0 if arcing_duration <= 0 else np.sum([U_arcing[i] * I_arcing[i] for i in range(arcing_duration)])
    
    # 燃弧电量
    arcing_electricity = 0 if arcing_duration <= 0 else np.sum([I_arcing[i] ** 2 for i in range(arcing_duration)])
    
    # 燃弧功率
    arcing_power = arcing_energy / (arcing_duration + 1)
    
    return arcing_range + [arcing_duration, arcing_energy, arcing_power]
```
