# 聚类分析

```python
from sklearn import preprocessing
import numpy as np
import pandas as pd
import os
import pickle
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## 1. 数据读取

### 获取燃弧特征

```python
root_data_path = '../data/'
processed_data_path = root_data_path + '1_processed_data/'
arcing_data_path = root_data_path + '2_arcing_data/'

tot_arcing_df = pd.DataFrame()
for machine_id in range(1, 11):
    pkl_file = open(processed_data_path + str(machine_id) + '.pkl', 'rb')
    machine_data = pickle.load(pkl_file)
    pkl_file.close()
    arcing_df = pd.read_csv(arcing_data_path + str(machine_id) + '.csv')
    arcing_df['machine_id'] = machine_id
    
    tot_arcing_df = tot_arcing_df.append(arcing_df, ignore_index = True)
```

### 特征归一化

```python
three_phase = ['A', 'B', 'C']
ftr_columns = [ t +'_arcing_' + ftr for t in three_phase for ftr in ['duration', 'energy', 'electricity', 'power']]
X = tot_arcing_df[ftr_columns]
scaler = preprocessing.StandardScaler().fit(X)
X_scaled = pd.DataFrame(scaler.transform(X), columns = ftr_columns)
```

```python
tot_arcing_df.head()
```

### 特征分布可视化

```python
import seaborn as sns  
def plot_dist(df, engine_num=None):
    '''plot all non trivial measurements and states'''
    
    cols = df.columns
    n_cols = min(len(cols), 4)
    n_rows = int(np.ceil(len(cols) / n_cols))
    
    sns.set()
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15,12))
    axes = axes.flatten()
    if engine_num != None: 
        fig.suptitle('distributions for engine #: {}'.format(engine_num))
        df_plot = df.loc[engine_num]
    else: 
        fig.suptitle('distributions for all engines')
        df_plot = df
    for col, ax in zip(cols, axes):
        ax=sns.distplot(df_plot[col], ax=ax, label=col)
        ax.legend(loc=1)
#         labels(col, "p", ax)
    return fig
```

```python
fig = plot_dist(tot_arcing_df)
```

```python
tot_arcing_df.describe()
```

## 2. 聚类分析

```python
from sklearn.cluster import KMeans
```

```python
# 抽取待聚类的特征
X_before_cluster = X_scaled[ftr_columns]
X_before_cluster.head()
```

### 聚类

```python
n_clusters = 5
kmeans = KMeans(n_clusters = n_clusters, n_init = 10, random_state = 1)
kmeans.fit(X_before_cluster)
```

```python
X_clustered = pd.DataFrame(X_before_cluster)
X_clustered_center = pd.DataFrame(kmeans.cluster_centers_, columns = X_clustered.columns)
X_clustered['label'] = kmeans.labels_
X_clustered_center['label'] = np.arange(n_clusters)
X_clustered = X_clustered.append(X_clustered_center)
```

```python
X_clustered.head()
```

### 统计各类燃弧出现的次数

```python
r1 = pd.Series(kmeans.labels_).value_counts()
r2 = pd.DataFrame(kmeans.cluster_centers_)
r = pd.concat([r2, r1], axis = 1)
r.columns = list(r2.columns) + [u'类别数']
r
```

### 可视化聚类结果

```python
from sklearn.manifold import TSNE
```

```python
#用TSNE进行数据降维并展示聚类结果
tsne = TSNE(random_state = 0)
tsne.fit_transform(X_clustered[X_clustered.columns.difference(['label'])]) #进行数据降维,并返回结果
X_tsne = pd.DataFrame(tsne.embedding_, index = X_clustered.index) #转换数据格式, index = X_clustered['label']
```

```python
X_tsne.shape, X_clustered.shape
```

```python
X_tsne['label'] = X_clustered['label']
```

```python
plt.figure(figsize=(15, 10))
for i in range(n_clusters):
    d = X_tsne[X_tsne['label'] == i]     #找出聚类类别为0的数据对应的降维结果
    #print(len(d))
    #print(d)
    plt.scatter(d[0], d[1])
d = X_tsne.tail(n_clusters)
plt.scatter(d[0], d[1], c='red', s=150, marker='*')
plt.show()
```

### 保存聚类结果

```python
tot_arcing_df['label'] = X_clustered[:-5]['label']
res_cols = tot_arcing_df.columns.tolist() + ['operation_times']
#['machine_id', 'operation_times', 'label','A_arcing_duration', 'A_arcing_energy', 'A_arcing_electricity', 'A_arcing_power', 'B_arcing_duration', 'B_arcing_energy', 'B_arcing_electricity', 'B_arcing_power', 'C_arcing_duration', 'C_arcing_energy', 'C_arcing_electricity', 'C_arcing_power']
res_df = pd.DataFrame([], columns = res_cols)

res_group = tot_arcing_df.groupby('machine_id')
for i in range(1, 11):
    tmp_df = res_group.get_group(i).reset_index(drop = True)
    tmp_df['operation_times'] = tmp_df.index
    #print(tmp_df.head())
    res_df = res_df.append(tmp_df, ignore_index = True)
res_df.head()
```

```python
res_path = '../res/'
if not os.path.exists(res_path):
    os.makedirs(res_path)
res_df.to_csv(res_path + '2_cluster_' + str(n_clusters) + '.csv', index = False, header = True)
```
